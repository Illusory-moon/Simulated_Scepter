```

from matplotlib import pyplot as plt
import cv2
from importing import load_img
def match_numbers_in_region(or_image):
    or_image = or_image[840:861, 1689:1704].copy()
    
    # 提取白色区域
    gray = cv2.cvtColor(or_image, cv2.COLOR_BGR2GRAY) if len(or_image.shape) == 3 else or_image
    mask = cv2.inRange(gray, 200, 255)
    white_region = cv2.bitwise_and(or_image, or_image, mask=mask)
    cv2.imwrite("0.png",white_region)
    plt.imshow(white_region, cmap='gray')
    plt.show()




```

```
load_img()
image = cv2.imread("20260404_233003.png")
results = match_numbers_in_region(image)
```
